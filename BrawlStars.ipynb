```python
from bs4 import BeautifulSoup
import requests

for brawler in brawlerArr:
  page = requests.get(f"https://brawlstars.fandom.com/wiki/{brawler}")
  soup = BeautifulSoup(page.content, "html.parser")
  image_url = soup.find_all(class_="pi-image-thumbnail")[0].get('src')
  print(image_url.split("/")[7].replace("_Skin-Default",""))
  response = requests.get(image_url)

  # Check if the request was successful
  if response.status_code == 200:
      # Define the path to store the image in Google Drive
      file_path = f'/content/drive/My Drive/Brawl Stars/{image_url.split("/")[7].replace("_Skin-Default","")}'

      # Save the image to Google Drive
      with open(file_path, 'wb') as f:
          f.write(response.content)
      print("Image saved to Google Drive!")
  else:
      print("Failed to retrieve the image.")
```
